```python
# import matplotlib
# matplotlib.use("agg")
import numpy as np
import matplotlib.pyplot as plt
```

## Basic Plotting

### Using pyplot to do some basic plotting

```python
# x_curve = 500 evenly spaced values between -2pi and +2pi
# y_curve = sin curve with personality of the values in x_curve

x_curve = np.linspace(-2 * np.pi, 2 * np.pi, 500)
y_curve = np.sinc(x_curve)
```

```python
plt.figure() # initialize figure
plt.plot(x_curve, y_curve) # draw curve described by x_curve and y_curve
plt.show() # display on screen
```

```python
x_scatter = np.linspace(-1, 1) # generate evenly spaced values between -1 and 1
y_scatter = x_scatter + np.random.normal(size=x_scatter.size) # similar y values with some gaussian distributed noise in them (zero mean and unit variance)
```

```python
plt.figure()
plt.scatter(x_scatter, y_scatter)
plt.show()
```

## Axes, backends

> A Figure is exactly what it sounds like -- a rectangular blank canvas on which a plot is constructed. Figures hold Axes objects. Axes objects are where all the plotting action happens and you can have more than one Axes in a Figure.

`
fig = plt.figure() # initilialize a new figure
ax = fig.gca() # get the current Axes object
ax.plot(x_curve, y_curve)
`

```python
x = np.linspace(-6, 6, 500)
y = np.sinc(x)
```

## Change width

```python
fig = plt.figure()
ax = fig.gca()
ax.plot(x, y, linewidth=2)
```

## Change color

```python
fig = plt.figure()
ax = fig.gca()
ax.plot(x, y, color="pink", linewidth="20")
```

## Changing the style

```python
fig = plt.figure()
ax = fig.gca()
ax.plot(x, y, linewidth=2, linestyle='-.') # -- dashed, : dotted, -. dash-dotted
ax.plot(x, y + .1, linewidth=2)
ax.plot(x, y + .2, linewidth=1)
ax.plot(x, y + .3, linewidth=.5)
ax.plot(x, y + .4, linewidth=.2)
```
